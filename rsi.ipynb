```python
import pandas as pd
from pandas import ExcelWriter
from pandas import ExcelFile
import numpy as np
import openpyxl
import talib
import matplotlib.pyplot as plt
```

```python
def _retrive_data(_symbol,_begin,_end):
    #获取数据并按日期升序排序
    data= pd.read_excel('goog.xlsx',sheetname = 'goog',index_col=0).sort_index()
    # data = data.sort_index()
    return data[_begin:_end]
```

```python
d = _retrive_data("goog","2015-01-09","2015-06")
open = d["Open"].values
close = d["Close"].values
high = d["High"].values
low = d["Low"].values
datem = d.index[:]
```

```python
cd = close[1:]-close[:-1]
cd
```

```python
u,d = [],[]

for i in cd:
    _u,_d = 0 , 0
    if i>_u:
        _u = i
    else:
        _d = i
    u.append(_u)
    d.append(_d)
    
```

```python
us = talib.MA(np.asarray(u),14,matype = 0)
ds = talib.MA(np.asarray(d),14,matype = 0)
```
